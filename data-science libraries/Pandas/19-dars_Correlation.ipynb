```python
# Generate random data
np.random.seed(0)  # For reproducibility
current_year = 2024
ages = np.random.randint(18, 80, 10)
birth_years = current_year - ages

# Ensure birth years are realistic
birth_years = np.where(birth_years > current_year - 100, birth_years, current_year - 100)

# Generate heights and weights
heights = np.random.normal(170, 10, 10)  # Normal distribution around 170 cm with standard deviation of 10
weights = np.random.normal(70, 15, 10)   # Normal distribution around 70 kg with standard deviation of 15

# Create DataFrame
data = {
    'Current Year': [current_year] * 10,  # All rows have the same current year
    'Age': ages,
    'Birth Year': birth_years,
    'Height (cm)': heights.round(2),  # Rounding to 2 decimal places
    'Weight (kg)': weights.round(2)   # Rounding to 2 decimal places
}

df1 = pd.DataFrame(data)
```
